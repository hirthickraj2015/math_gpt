## Math Dataset Generation

```python
def evaluate_expression(expr):
    try:
        result = eval(expr)
        if isinstance(result, float):
            if result.is_integer():
                return int(result)
            return round(result, 2)
        return result
    except:
        return None
```

```python
def generate_simple_expressions(num_samples):
    expressions = []
    ops = ['+', '-', '*', '//', '%']
    
    for _ in range(num_samples):
        a = random.randint(1, 100)
        b = random.randint(1, 100)
        op = random.choice(ops)
        
        if op == '//' and b == 0:
            b = random.randint(1, 100)
        if op == '%' and b == 0:
            b = random.randint(1, 100)
        
        expr = f"{a}{op}{b}"
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_two_operation_expressions(num_samples):
    expressions = []
    ops = ['+', '-', '*', '//', '%']
    
    for _ in range(num_samples):
        a = random.randint(1, 50)
        b = random.randint(1, 50)
        c = random.randint(1, 50)
        
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        
        if op1 == '//' and b == 0:
            b = random.randint(1, 50)
        if op2 == '//' and c == 0:
            c = random.randint(1, 50)
        if op1 == '%' and b == 0:
            b = random.randint(1, 50)
        if op2 == '%' and c == 0:
            c = random.randint(1, 50)
        
        expr = f"{a}{op1}{b}{op2}{c}"
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_parentheses_expressions(num_samples):
    expressions = []
    ops = ['+', '-', '*', '//', '%']
    
    for _ in range(num_samples):
        a = random.randint(1, 30)
        b = random.randint(1, 30)
        c = random.randint(1, 30)
        
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        
        if op1 == '//' and b == 0:
            b = random.randint(1, 30)
        if op2 == '//' and c == 0:
            c = random.randint(1, 30)
        if op1 == '%' and b == 0:
            b = random.randint(1, 30)
        if op2 == '%' and c == 0:
            c = random.randint(1, 30)
        
        pattern = random.choice(['left', 'right'])
        
        if pattern == 'left':
            expr = f"({a}{op1}{b}){op2}{c}"
        else:
            expr = f"{a}{op1}({b}{op2}{c})"
        
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_nested_expressions(num_samples):
    expressions = []
    ops = ['+', '-', '*', '//', '%']
    
    for _ in range(num_samples):
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        c = random.randint(1, 20)
        d = random.randint(1, 20)
        
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        op3 = random.choice(ops)
        
        if op1 == '//' and b == 0:
            b = random.randint(1, 20)
        if op2 == '//' and c == 0:
            c = random.randint(1, 20)
        if op3 == '//' and d == 0:
            d = random.randint(1, 20)
        if op1 == '%' and b == 0:
            b = random.randint(1, 20)
        if op2 == '%' and c == 0:
            c = random.randint(1, 20)
        if op3 == '%' and d == 0:
            d = random.randint(1, 20)
        
        pattern = random.choice(['nested_left', 'nested_right', 'double_paren'])
        
        if pattern == 'nested_left':
            expr = f"(({a}{op1}{b}){op2}{c}){op3}{d}"
        elif pattern == 'nested_right':
            expr = f"{a}{op1}({b}{op2}({c}{op3}{d}))"
        else:
            expr = f"({a}{op1}{b}){op2}({c}{op3}{d})"
        
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_large_number_expressions(num_samples):
    expressions = []
    ops = ['+', '-', '*', '//', '%']
    
    for _ in range(num_samples):
        a = random.randint(100, 1000)
        b = random.randint(1, 100)
        op = random.choice(ops)
        
        if op == '//' and b == 0:
            b = random.randint(1, 100)
        if op == '%' and b == 0:
            b = random.randint(1, 100)
        
        expr = f"{a}{op}{b}"
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_complex_bodmas_expressions(num_samples):
    expressions = []
    ops_high = ['*', '//']
    ops_low = ['+', '-']
    
    for _ in range(num_samples):
        a = random.randint(2, 25)
        b = random.randint(2, 25)
        c = random.randint(2, 25)
        d = random.randint(2, 25)
        
        op1 = random.choice(ops_high)
        op2 = random.choice(ops_low)
        op3 = random.choice(ops_high)
        
        if op1 == '//' and b == 0:
            b = random.randint(2, 25)
        if op3 == '//' and d == 0:
            d = random.randint(2, 25)
        
        expr = f"{a}{op1}{b}{op2}{c}{op3}{d}"
        result = evaluate_expression(expr)
        
        if result is not None:
            expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def save_dataset(filepath, expressions):
    import os
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    with open(filepath, 'w') as f:
        for expr in expressions:
            f.write(expr + '\n')
```

```python
print("Generating math dataset...")

all_math_expressions = []

print("Generating simple expressions...")
all_math_expressions.extend(generate_simple_expressions(30000))

print("Generating two operation expressions...")
all_math_expressions.extend(generate_two_operation_expressions(25000))

print("Generating parentheses expressions...")
all_math_expressions.extend(generate_parentheses_expressions(20000))

print("Generating nested expressions...")
all_math_expressions.extend(generate_nested_expressions(15000))

print("Generating large number expressions...")
all_math_expressions.extend(generate_large_number_expressions(15000))

print("Generating complex BODMAS expressions...")
all_math_expressions.extend(generate_complex_bodmas_expressions(15000))

# Shuffle to mix different types
random.shuffle(all_math_expressions)

# Split: 90% training, 10% testing
split_idx = int(0.9 * len(all_math_expressions))
math_train = all_math_expressions[:split_idx]
math_test = all_math_expressions[split_idx:]

# Save to new directory structure
save_dataset('dataset/math/training/math_train.txt', math_train)
save_dataset('dataset/math/testing/math_test.txt', math_test)

print(f"\nMath dataset generation complete!")
print(f"Total expressions: {len(all_math_expressions)}")
print(f"Training samples: {len(math_train)}")
print(f"Testing samples: {len(math_test)}")
print(f"\nSaved to:")
print(f"  Training: dataset/math/training/math_train.txt")
print(f"  Testing: dataset/math/testing/math_test.txt")
print("\nSample expressions:")
for i in range(10):
    print(f"  {all_math_expressions[i]}")
```

## Boolean Logic Dataset Generation

```python
def generate_simple_boolean_expressions(num_samples):
    expressions = []
    ops = ['AND', 'OR', 'XOR']
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        b = random.choice(values)
        op = random.choice(ops)
        
        expr = f"{a} {op} {b}"
        
        if op == 'AND':
            result = eval(f"{a} and {b}")
        elif op == 'OR':
            result = eval(f"{a} or {b}")
        else:
            result = (a == 'True') != (b == 'True')
        
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_not_expressions(num_samples):
    expressions = []
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        expr = f"NOT {a}"
        result = eval(f"not {a}")
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_two_boolean_operations(num_samples):
    expressions = []
    ops = ['AND', 'OR', 'XOR']
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        b = random.choice(values)
        c = random.choice(values)
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        
        expr = f"{a} {op1} {b} {op2} {c}"
        
        temp_a = a
        temp_b = b
        temp_c = c
        
        if op1 == 'AND':
            temp = eval(f"{temp_a} and {temp_b}")
        elif op1 == 'OR':
            temp = eval(f"{temp_a} or {temp_b}")
        else:
            temp = (temp_a == 'True') != (temp_b == 'True')
        
        if op2 == 'AND':
            result = eval(f"{temp} and {temp_c}")
        elif op2 == 'OR':
            result = eval(f"{temp} or {temp_c}")
        else:
            result = temp != (temp_c == 'True')
        
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_boolean_with_parentheses(num_samples):
    expressions = []
    ops = ['AND', 'OR', 'XOR']
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        b = random.choice(values)
        c = random.choice(values)
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        
        pattern = random.choice(['left', 'right'])
        
        if pattern == 'left':
            expr = f"({a} {op1} {b}) {op2} {c}"
            
            if op1 == 'AND':
                temp = eval(f"{a} and {b}")
            elif op1 == 'OR':
                temp = eval(f"{a} or {b}")
            else:
                temp = (a == 'True') != (b == 'True')
            
            if op2 == 'AND':
                result = eval(f"{temp} and {c}")
            elif op2 == 'OR':
                result = eval(f"{temp} or {c}")
            else:
                result = temp != (c == 'True')
        else:
            expr = f"{a} {op1} ({b} {op2} {c})"
            
            if op2 == 'AND':
                temp = eval(f"{b} and {c}")
            elif op2 == 'OR':
                temp = eval(f"{b} or {c}")
            else:
                temp = (b == 'True') != (c == 'True')
            
            if op1 == 'AND':
                result = eval(f"{a} and {temp}")
            elif op1 == 'OR':
                result = eval(f"{a} or {temp}")
            else:
                result = (a == 'True') != temp
        
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_boolean_with_not(num_samples):
    expressions = []
    ops = ['AND', 'OR', 'XOR']
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        b = random.choice(values)
        op = random.choice(ops)
        
        pattern = random.choice(['not_left', 'not_right', 'not_both'])
        
        if pattern == 'not_left':
            expr = f"NOT {a} {op} {b}"
            temp_a = eval(f"not {a}")
            temp_b = b
        elif pattern == 'not_right':
            expr = f"{a} {op} NOT {b}"
            temp_a = a
            temp_b = eval(f"not {b}")
        else:
            expr = f"NOT {a} {op} NOT {b}"
            temp_a = eval(f"not {a}")
            temp_b = eval(f"not {b}")
        
        if op == 'AND':
            result = eval(f"{temp_a} and {temp_b}")
        elif op == 'OR':
            result = eval(f"{temp_a} or {temp_b}")
        else:
            result = temp_a != temp_b
        
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
def generate_complex_boolean_expressions(num_samples):
    expressions = []
    ops = ['AND', 'OR', 'XOR']
    values = ['True', 'False']
    
    for _ in range(num_samples):
        a = random.choice(values)
        b = random.choice(values)
        c = random.choice(values)
        d = random.choice(values)
        op1 = random.choice(ops)
        op2 = random.choice(ops)
        op3 = random.choice(ops)
        
        pattern = random.choice(['nested', 'double_paren'])
        
        if pattern == 'nested':
            expr = f"({a} {op1} ({b} {op2} {c})) {op3} {d}"
            
            if op2 == 'AND':
                temp1 = eval(f"{b} and {c}")
            elif op2 == 'OR':
                temp1 = eval(f"{b} or {c}")
            else:
                temp1 = (b == 'True') != (c == 'True')
            
            if op1 == 'AND':
                temp2 = eval(f"{a} and {temp1}")
            elif op1 == 'OR':
                temp2 = eval(f"{a} or {temp1}")
            else:
                temp2 = (a == 'True') != temp1
            
            if op3 == 'AND':
                result = eval(f"{temp2} and {d}")
            elif op3 == 'OR':
                result = eval(f"{temp2} or {d}")
            else:
                result = temp2 != (d == 'True')
        else:
            expr = f"({a} {op1} {b}) {op2} ({c} {op3} {d})"
            
            if op1 == 'AND':
                temp1 = eval(f"{a} and {b}")
            elif op1 == 'OR':
                temp1 = eval(f"{a} or {b}")
            else:
                temp1 = (a == 'True') != (b == 'True')
            
            if op3 == 'AND':
                temp2 = eval(f"{c} and {d}")
            elif op3 == 'OR':
                temp2 = eval(f"{c} or {d}")
            else:
                temp2 = (c == 'True') != (d == 'True')
            
            if op2 == 'AND':
                result = eval(f"{temp1} and {temp2}")
            elif op2 == 'OR':
                result = eval(f"{temp1} or {temp2}")
            else:
                result = temp1 != temp2
        
        expressions.append(f"{expr}={result}")
    
    return expressions
```

```python
print("Generating boolean logic dataset...")

all_boolean_expressions = []

print("Generating simple boolean expressions...")
all_boolean_expressions.extend(generate_simple_boolean_expressions(25000))

print("Generating NOT expressions...")
all_boolean_expressions.extend(generate_not_expressions(10000))

print("Generating two operation expressions...")
all_boolean_expressions.extend(generate_two_boolean_operations(20000))

print("Generating expressions with parentheses...")
all_boolean_expressions.extend(generate_boolean_with_parentheses(20000))

print("Generating expressions with NOT...")
all_boolean_expressions.extend(generate_boolean_with_not(15000))

print("Generating complex expressions...")
all_boolean_expressions.extend(generate_complex_boolean_expressions(15000))

# Shuffle to mix different types
random.shuffle(all_boolean_expressions)

# Split: 90% training, 10% testing
split_idx = int(0.9 * len(all_boolean_expressions))
boolean_train = all_boolean_expressions[:split_idx]
boolean_test = all_boolean_expressions[split_idx:]

# Save to new directory structure
save_dataset('dataset/boolean/training/boolean_train.txt', boolean_train)
save_dataset('dataset/boolean/testing/boolean_test.txt', boolean_test)

print(f"\nBoolean dataset generation complete!")
print(f"Total expressions: {len(all_boolean_expressions)}")
print(f"Training samples: {len(boolean_train)}")
print(f"Testing samples: {len(boolean_test)}")
print(f"\nSaved to:")
print(f"  Training: dataset/boolean/training/boolean_train.txt")
print(f"  Testing: dataset/boolean/testing/boolean_test.txt")
print("\nSample expressions:")
for i in range(10):
    print(f"  {all_boolean_expressions[i]}")
```
